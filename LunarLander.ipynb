```python
import gymnasium

from huggingface_sb3 import load_from_hub, package_to_hub
from huggingface_hub import notebook_login

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

```python
#gymnasium.make makes the environment
# as for every observation the enviroment needs to reset to its inital position obsserbvation = env.reset()
#we need to take a random action now hece, env.stoep(action)
#observation: The new state (st+1)
#reward: The reward we get after executing the action
#terminated: Indicates if the episode terminated (agent reach the terminal state)
#truncated: Introduced with this new version, it indicates a timelimit or if an agent go out of bounds of the environment for instance.
#info: A dictionary that provides additional information (depends on the environment)
```

```python
import gymnasium as gym

#creating the environment called lunarlander-v2

env = gym.make("LunarLander-v2")
#reseting the env

observation , info = env.reset()

for _ in range(20):
  #take a random action

  action = env.action_space.sample()
  print("Action taken:", action)

  #do this action in the env and get next stage , reward , terminated , truncated and info

  observation, reward, terminated, truncated, info = env.step(action)

  #if the game is terminated (in our case landed or crashed) or truncated - timeout
  if terminated or truncated:
    #reset the env
    print("environment is reset")
    observation, info = env.reset()

  env.close()
```

```python
#setting up the environment

env = gym.make("LunarLander-v2")
env.reset()
print("_______OBSERVATION SPACE_______")
print("Observation Space Shape", env.observation_space.shape)
print("Sample Observation", env.observation_space.sample())
```

```python
#creating the environment

env = make_vec_env('LunarLander-v2', n_envs = 16)

```

```python
model = PPO(
    policy = 'MlpPolicy',#multi layer perceptron policy
    env = env,
    n_steps = 1024,
    batch_size = 64,
    n_epochs = 4,
    gamma = 0.999,
    gae_lambda = 0.98,
    ent_coef = 0.01,
    verbose = 1
)
```

```python
#run the model for 1000000 one million times that is
model.learn(total_timesteps= 1000000)

model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

```python
eval_env = Monitor(gym.make("LunarLander-v2", render_mode ='rgb_array'))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes = 10,deterministic = True)
print(f"mean_reward = {mean_reward: 2f} +/- {std_reward:2f}")
```
